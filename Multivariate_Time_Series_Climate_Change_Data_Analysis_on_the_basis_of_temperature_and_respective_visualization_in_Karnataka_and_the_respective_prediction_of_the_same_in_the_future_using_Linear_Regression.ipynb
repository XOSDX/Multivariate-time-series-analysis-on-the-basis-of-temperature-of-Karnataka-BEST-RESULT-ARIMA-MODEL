**EVALUATION USING LINEAR REGRESSION MODEL:**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
KDXS=KDX.drop('Period',axis=1)
KDXS
```

```python
# Separate the 'PC' and 'Mean' columns into X (features) and y (target)
X = KDXS.drop(['1901'],axis=1)
y = KDXS.drop('Mean',axis=1)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create a linear regression model
model = LinearRegression()
```

```python
X
```

```python
y
```

```python
model.fit(X_train,y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
# Print the coefficients
print(f'Coefficients: {model.coef_}, {model.intercept_}')
```

```python
# Plot the results
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred)
plt.xlabel('Year')
```
